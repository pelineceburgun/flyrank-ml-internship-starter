## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

## 3. 🔧 Your turn

Pick **one** of these and write a few lines below. This is your Week-1 discovery — you'll reference it in your Week-1 research-question write-up (on the InternHQ board).

- Redo Discovery A but only for pages with `impressions_90d > 0` — does the correlation change?
- In Discovery B, which `content_type` has the worst CTR *at the same position tier*?
- Find another belief to test: does `content_age_days` relate to `trend_direction`? Does `avg_position` relate to `ctr`?

**Rules:** describe what you observe as *observed / directional* — never "I proved Google's algorithm." Keep client data out of anything you publish.

```python
# Your discovery here

# Discovery A (refined): Does the search_volume–impressions correlation
# change when we filter to pages that actually got traffic?

df_with_traffic = df[df["impressions_90d"] > 0]
corr_all    = df["search_volume"].corr(df["impressions_90d"])
corr_active = df_with_traffic["search_volume"].corr(df_with_traffic["impressions_90d"])

print(f"Rows with impressions > 0  : {len(df_with_traffic):,} / {len(df):,}")
print(f"Correlation (all rows)     : {corr_all:.3f}")
print(f"Correlation (traffic > 0)  : {corr_active:.3f}")
print()
if abs(corr_active) < 0.15:
    print("Still near zero → even among pages Google is already showing,")
    print("keyword search volume does not reliably predict how many impressions a page earns.")
else:
    print("The correlation rises when zero-traffic pages are removed — but still weak.")
    print("This suggests zero-traffic pages are a distinct class (not just low performers),")
    print("not simply noise dragging down the average.")

# ── Bonus Discovery: avg_position vs ctr ──────────────────────────────────
# Hypothesis: lower (worse) avg_position → lower CTR, but is it linear?

import matplotlib.pyplot as plt
import numpy as np

visible = df[df["impressions_90d"] >= 100].copy()

# Bin avg_position into integer buckets 1-20, treat >20 as 21
visible["pos_bucket"] = visible["avg_position"].clip(1, 21).round(0)
ctr_by_pos = visible.groupby("pos_bucket")["ctr"].mean()

print("\n── avg_position vs mean CTR (pages with ≥100 impressions) ──")
print(ctr_by_pos.round(4).to_string())

ctr_by_pos.plot(
    kind="line", marker="o",
    title="CTR drops steeply after position 3 — not gradual",
    xlabel="avg_position (1 = top)", ylabel="Mean CTR"
)
plt.tight_layout()
plt.show()

print()
print("Observation: the CTR drop is not linear — there is a cliff around position 3-5.")
print("Jumping from position 10 → 5 matters far less than from position 5 → 2.")
print("Implication: refresh priority should not just target 'declining' pages,")
print("but specifically those close to the position-3 threshold where CTR gains are steepest.")
```

```python
# ── REVISED INTERPRETATION ─────────────────────────────────────────────────

print("""
Key findings:

1. Search volume → impressions correlation: 0.001 (all rows) and 0.001 (traffic > 0)
   → Filtering zero-traffic pages changes nothing.
   → Search volume is not a proxy for actual impressions, full stop.

2. CTR vs avg_position — unexpected shape:
   - Peak CTR is at position 3-4 (~0.42-0.44), NOT position 1 (0.12)
   - Position 1 underperforms positions 2-6 significantly
   - After position 5, CTR declines gradually to ~0.23 at positions 16-20
   - This is NOT the classic monotonic "higher rank = higher CTR" curve

3. Practical implication:
   - Refresh priority should target pages in positions 5-10 that are close
     to the 3-5 band — that is where the CTR reward is steepest
   - Pages already AT position 1 may have a structural CTR ceiling
     (ads, featured snippets, or high-competition keyword mix)
   - Limitation: avg_position is a per-page average across all keywords;
     a page ranking 1st for some queries and 20th for others may distort
     this analysis. A per-query-level dataset would be more reliable.
""")
```
